```python
import pickle
import svm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\firda\OneDrive\Documents\Project\Python\capstone\Data_Fix.csv")
df = df.iloc[:, 1:]
df
```

```python
bins = [0, 2, 2.5, 3, 3.5, 4]
labels = ['<2.0', '2.0-2.5', '2.5-3.0', '3.0-3.5', '3.5-4.0']
df['IPK_Range'] = pd.cut(df['IPK'], bins=bins, labels=labels, include_lowest=True)
df['IPK'] = df['IPK_Range']
df.drop('IPK_Range', axis=1, inplace=True)
print(df)
```

```python
df['Prestasi'] = df['Prestasi'].str.extract(r'(Juara \d)').fillna('')
df
```

```python
bins = [0, 300, 400, 500, 600, 700]
labels = ['<300', '300-400', '400-500', '500-600', '600-700']
df['Toefl_Range'] = pd.cut(df['Toefl'], bins=bins, labels=labels, include_lowest=True)
df['Toefl'] = df['Toefl_Range']
df.drop('Toefl_Range', axis=1, inplace=True)
print(df)
```

```python
df['IPK'] = df['IPK'].replace({'Tidak ada' : 0, 'Ada' : 1})
df['Status Orang Tua Siswa'] = df['Status Orang Tua Siswa'].replace({'Lengkap' : 0.5, 'Yatim Piatu' : 1, 'Yatim / Piatu' : 0.75})
df['Penghasilan Orang Tua'] = df['Penghasilan Orang Tua'].replace({'Diatas 1 Juta' : 0.5, 'Dibawah 1 Juta' : 0.75, 'Tidak Berpenghasilan' : 1})
df['Data Dinas Sosial'] = df['Data Dinas Sosial'].replace({'Normal' : 0.5, 'Miskin' : 1, 'Rentang Miskin' : 0.75})
df['Target'] = df['Target'].replace({'Ya' : 1, 'Tidak' : 0})

df['Status Orang Tua Siswa'] = df['Status Orang Tua Siswa'].astype(float)
df['Penghasilan Orang Tua'] = df['Penghasilan Orang Tua'].astype(float)
df['Data Dinas Sosial'] = df['Data Dinas Sosial'].astype(float)
df
```
